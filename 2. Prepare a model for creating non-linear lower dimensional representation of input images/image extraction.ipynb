```python
import os
from pathlib import Path
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
p = Path("./cityscapes_data/train/")
```

```python
img_arr_train = []
for img_path in p.glob("*.jpg"):
    img = image.load_img(img_path)
    img_arr_train.append(img)
```

```python
def drawImg(img):
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
x_train = []
y_train = []
for i in range(len(img_arr_train)):
    h,w = img_arr_train[i].size
    x_train.append(img_arr_train[i].crop((0,0,h/2,w)).resize((256,256)))
    y_train.append(img_arr_train[i].crop((h/2,0,h,w)).resize((256,256)))
```

```python
for i in range(10):
    drawImg(x_train[i])
```

```python
for i in range(10):
    drawImg(y_train[i])
```

```python
for i in range(len(x_train)):
    x_train[i].save("./cityscapes_data/x_train/"+str(i)+".jpg")
```

```python
for i in range(len(x_train)):
    y_train[i].save("./cityscapes_data/y_train/"+str(i)+".jpg")
```
